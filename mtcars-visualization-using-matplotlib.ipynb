```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mtcars = pd.read_csv("../input/mtcars/mtcars.csv")
```

So, am feature / predictor has 2 categories:

*   “AT” for Automatic Transmission
*   “MT” for Manual Transmission

```python
mtcars["am"].unique()
```

```python
x = mtcars["am"].value_counts()
x
```

```python
mtcars.groupby('gear').am.value_counts()
```

```python
#Bar plot
plt.figure(figsize=(10,7), facecolor='black')
with plt.style.context('dark_background'):
   ax = pd.crosstab(mtcars.gear,mtcars.am).plot(kind="bar", figsize=(10,8))
plt.xticks(rotation=0)
for p in ax.patches:
        ax.text(p.get_x() + p.get_width()/2 ,p.get_height(),  
                "%.f" % p.get_height(),
                color='black', rotation="horizontal", fontsize=12, horizontalalignment="center", verticalalignment="center", 
                bbox=dict(boxstyle="Round", fc="w"))
plt.grid()
plt.show()
```

```python
mtcars["model"].unique
```

```python
#Histogram
mtcars.hist(figsize=(15,15))
plt.show()
```

```python
plt.figure(figsize=(15,10),facecolor = "lavender")

plt.subplot(3,3,1,xlabel='disp',ylabel='mpg')
plt.scatter(mtcars['disp'],mtcars['mpg'],c='green' )
plt.plot(mtcars['disp'],mtcars['mpg'], '-o')

plt.subplot(3,3,2,xlabel='disp',ylabel='hp')
plt.scatter(mtcars['disp'],mtcars['hp'] ,marker = "^" ,c ="peru")
plt.plot(mtcars['disp'],mtcars['hp'], '--' , c="gold")

plt.subplot(3,3,3, xlabel='wt',ylabel='mpg')
plt.scatter(mtcars['wt'],mtcars['mpg'] , c = "green")
plt.plot(mtcars['wt'],mtcars['mpg'], '--' , c = "#4CAF50")

plt.subplot(3,3,4, xlabel='wt',ylabel='disp')
plt.scatter(mtcars['wt'],mtcars['disp'],c='purple')
plt.plot(mtcars['wt'],mtcars['disp'], '--' , c = 'm')

plt.subplot(3,3,5, xlabel='hp',ylabel='qsec')
plt.scatter(mtcars['hp'],mtcars['qsec'],c='skyblue')
plt.plot(mtcars['hp'],mtcars['qsec'], '-*' , c ="c")

plt.subplot(3,3,6, xlabel='hp',ylabel='mpg')
plt.scatter(mtcars['hp'],mtcars['mpg'] , c ="yellowgreen")
plt.plot(mtcars['hp'],mtcars['mpg'], '--' , c = "darkolivegreen")

plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.6)
plt.show()
```

```python
#Correlation matrix
Correlation = mtcars.corr()
Correlation.style.background_gradient()
```

```python
#line chart
plt.figure(figsize=(10,7))
plt.axes(facecolor='black')
plt.plot(mtcars["disp"] , label = "disp" , linewidth =3)
plt.plot(mtcars["hp"] , label = "hp", linewidth =3)
plt.legend(loc='upper right')
plt.show()
```

```python
#line chart
plt.figure(figsize=(10,7) )
plt.axes(facecolor='black')
plt.plot(mtcars["wt"], label = "wt" , linewidth =3)
plt.plot(mtcars["cyl"], label = "cyl" , linewidth =3)
plt.legend(loc='upper right')
plt.show()
```

```python
# bar plot between 2 different categories 
with plt.style.context('dark_background'):
  pd.crosstab(mtcars.gear,mtcars.cyl).plot(kind="bar", figsize=(10,8),  stacked = True  , edgecolor = "white",linewidth= 2.5)
plt.xticks(rotation=0)
plt.show()
```

```python
mtcars["gear"].value_counts()
```

```python
#Pie plot
plt.figure(figsize=(10,7)).patch.set_facecolor('black')
mtcars["gear"].value_counts()
myexplode = [0.1, 0.1, 0.1]
mycolor = ["c" , "peru" ,"#4CAF50"]
plt.ylabel('gear', color='white')
mtcars.gear.value_counts().plot(kind="pie" , startangle = 90 ,  explode = myexplode , shadow = True , colors = mycolor , autopct='%.0f%%' ,wedgeprops = {"edgecolor" : "white",
                      'linewidth': 2,
                      'antialiased': True}) #for decimal value =autopct='%.2f%%'
plt.legend(title = "MT-Cars Gear") 
plt.show()
```

```python
#scatter plot with line chart
plt.figure(figsize=(15,5) , facecolor='skyblue' )
plt.axes(facecolor='skyblue').spines["left"].set_color('black')
plt.tick_params( colors='black', which='both' )
#[‘.’,’o’,’v’,’^’,’>’,'<‘,’s’,’p’,’*’,’h’,’H’,’D’,’d’,’1′,”,”] = markers
plt.scatter(mtcars["mpg"],mtcars["qsec"], color = "black" ,cmap='Spectral')## scatter plot of two variables
plt.xlabel('mpg',color='black',size=10 , fontweight ="bold")
plt.ylabel('qsec',color='black',size=10 ,fontweight ="bold")
plt.title('Scatterplot', color='skyblue')


# Now add on a line with a fixed slope of 0.03
#slope = 0.03

# A line with a fixed slope can intercept the axis
# anywhere so we're going to have it go through 0,0
x_0 = 10
y_0 = 16

# And we'll have the line stop at x = 34
x_1 = 34
y_1 = 22  #slope*(x_1) + y_0

# Draw these two points with big triangles to make it clear
# where they lie
plt.scatter([x_0, x_1], [y_0, y_1], c='r')

# And now connect them
plt.plot([x_0, x_1], [y_0, y_1], c='r')

plt.legend()
plt.grid()
plt.show()
```

```python
mtcars= mtcars.sort_values('disp',ascending=True)
mtcars
```

```python
#creates histogram with 5bins and colours filled init.
plt.figure(figsize = (16,12))
plt.axes(facecolor='black')
plt.barh(mtcars['model'], mtcars['disp'] ,facecolor ='#99ff99',edgecolor ="red" ,linewidth = 1 )

for index, value in enumerate(mtcars["disp"]):
    plt.text(value,index , str(value),
            color = 'white', fontweight = 'bold' , size = 15) # ha='center'

plt.grid()
plt.show()
```

```python
# vertical Boxplot
plt.axes(facecolor='black')
c = "lime"
plt.boxplot(mtcars['mpg'],vert =False , boxprops=dict( color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=c),widths = 2 )
plt.ylabel("MPG")
plt.xlabel("Boxplot")
plt.title("Boxplot")  
plt.show()
```
